# Handling Null values

```python
data_new = {
    'Day':['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
    'unit_sold':[98.00,34.89,73.54,49.34,48.9,np.nan,np.nan]
}
null_test =pd.DataFrame(data_new)
null_test.to_feather('null_test.feather')
null_test
```

```python
null_test = pd.read_feather('null_test.feather')
null_test.head()
null_test.dropna()#droping null values cloumns
#filling null values with zero's
null_test.fillna(0)#that means assumming no units are sold on weekends
#filling null values with mean value of unit_sold
null_test.fillna(null_test['unit_sold'].mean())
#filling null values by interpolate values of that columns
null_test['unit_sold'].fillna(null_test['unit_sold'].interpolate(),inplace=True)
null_test
```

# Aggregation,GroupBY and Pivot Table

```python
pokemon.head()
grass = pokemon[pokemon['Type 1'] == 'Grass']['HP'].value_counts().sort_values(ascending=True)
grass
fire = pokemon[pokemon['Type 1'] == 'Fire'] ['Legendary'].value_counts()
fire
pokemon.head()
```

```python
pokemon.groupby(['Type 1']).agg({'HP':['mean','max','min'],'Attack':['mean','max','min']})
pokemon.groupby(['Type 1','Type 2']).agg({'HP':['mean','max','min'],'Attack':['mean','max','min']})
pokemon.groupby(['Type 1']).agg({'HP':['mean','max','min'],'Attack':['mean','max','min']}).rank()
pokemon.groupby(['Type 1','Type 2']).agg({'HP':['mean','max','min'],'Attack':['mean','max','min']}).rank()

```

```python
pivot = pokemon.pivot_table(columns='Type 1',index='Type 2',values='HP')
pivot = pokemon.pivot_table(columns='Type 1',index='Type 2',values='HP',aggfunc='sum')
pivot
```

```python
grouped = pokemon.groupby(['Type 1']).agg({'HP':['mean','max','min'],'Attack':['mean','max','min']})
# grouped['Attack_max'].rank() it gives an key_error
grouped['Attack']['max'].rank()#its working simliar version of this is
grouped.columns = ['_'.join(col).strip('_') for col in grouped.columns.values]
(grouped['HP_mean']+grouped['Attack_mean']).rank()
```

ðŸŽ¯ Challenge 3: Strongest Dual-Type (Skill Rate)
ðŸ‘‰ Task: For each dual-type (not NaN in Type 2), calculate average skill_rate and get the top 5.


```python
pokemon_multi['Skill Rate'] = pokemon_multi['Attack'] + pokemon_multi['Defense'] + pokemon_multi['Sp. Atk'] + pokemon_multi['Sp. Def'] + pokemon_multi['Speed']
pokemon_multi.loc[pokemon_multi.index.get_level_values('Type 2').notna()].groupby(level=[0,1]).agg({'Skill Rate':'mean'}).sort_values('Skill Rate',ascending=False).head(5)
```
